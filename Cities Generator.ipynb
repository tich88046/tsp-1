```python
import pandas as pd
import numpy as np
n=100
cities=np.empty((n,3))
cities[:,0]=np.array(range(n))+1
cities[:,1:]=np.random.rand(n,2)*10
cities={'id':cities[:,0],'x':cities[:,1],'y':cities[:,2]}
cities=pd.DataFrame(cities)
cities.head(10)
```

```python
cities.to_csv(path_or_buf="./data/cities.csv", sep=" ",index=False,mode='w+')
```
