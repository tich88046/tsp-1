```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('./data/cities.csv', sep=' ')
print(data)
```

```python
#Copied and modified from: 
#https://stackoverflow.com/questions/21352580/matplotlib-plotting-numerous-disconnected-line-segments-with-different-colors
#https://matplotlib.org/2.1.1/gallery/shapes_and_collections/line_collection.html
#https://stackoverflow.com/questions/14432557/matplotlib-scatter-plot-with-different-text-at-each-data-point
import numpy as np
import pylab as pl
from matplotlib import collections  as mc
def loadPath(path):
    i=0
    lines=np.zeros((path.size, 2, 2), float)
    c=np.zeros((path.size,4))
    #print(test)
    #print(lines)
    while(i<path.size):
        lines[i,0,0]=data.iloc[path[i]-1,1]
        lines[i,0,1]=data.iloc[path[i]-1,2]
        lines[i,1,0]=data.iloc[path[(i+1)%path.size]-1 ,1]
        lines[i,1,1]=data.iloc[path[(i+1)%path.size]-1 ,2]
        c[i]=((i+1)/path.size,0,0,1)
        i=i+1
    return lines,c

fig, ax = plt.subplots()
ax.scatter(data['x'],data['y'])
n=np.arange(data.shape[0])+1
for i, txt in enumerate(n):
    ax.annotate(txt, (data['x'][i], data['y'][i]))
    
#modify this to change path
s="87,19,62,58,45,75,14,11,81,93,42,37,98,12,52,21,25,33,9,51,86,15,50,40,48,5,24,41,88,16,34,83,22,66,44,91,95,65,30,80,43,36,57,32,89,77,29,71,18,63,53,100,54,74,59,56,31,70,7,84,69,23,1,67,68,60,94,73,3,90,82,4,2,72,6,76,47,20,49,27,39,78,79,85,13,8,96,38,55,17,92,10,35,64,28,61,26,97,46,99"
path=np.array(s.split(","),int)
lines,c=loadPath(path)
lc = mc.LineCollection(lines, colors=c, linewidths=2)
fig, ax = pl.subplots()
ax.add_collection(lc)
ax.autoscale()
ax.margins(0.1)
#ax.margins(0.1)
```
